```python
import pandas as pd
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df
```

დუპლიკატებისგან გაწმენდა

```python
df.drop_duplicates()
df
```

სვეტის ამოშლა

```python
df.drop(columns="Magnitude Error")
df
```

Cut ფუნქცია. დახარისხება სიღრმის პატარა, საშუალო და მაღალ დონეებად

```python
df['Depth_Categories'] = pd.cut(df['Depth'],
                                bins=[int(df['Depth'].min())-1, 40, 100, int(df['Depth'].max())+1],
                                labels=['low', 'medium', 'high'])

df
```

Reshaping, stack და unstack fუნქციები

```python
df_stacked = df.stack(level=0)
df_stacked
```

```python
df_stacked.unstack()
```
