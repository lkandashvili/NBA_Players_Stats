### Filtering

```python
filter_Lakers_Players_Not_From_USA = (data['team_abbreviation'] == 'LAL') & (data['country'] != 'USA')

filter_Lakers_Players_From_USA = (data['team_abbreviation'] == 'LAL') & (data['country'] == 'USA')
```

```python
df.loc[filter_Lakers_Players_Not_From_USA]
```

```python
df.loc[filter_Lakers_Players_From_USA]
```

### Sorting

```python
df.sort_values(by= ['draft_round', 'draft_year'], ascending= True)
```

### Mean height of non american Lakers players

```python
nonAmericanLakersMean = df.loc[filter_Lakers_Players_Not_From_USA, 'player_height'].mean()
nonAmericanLakersMean

```

### Mean height of american Lakers players

```python
americanLakersMean= df.loc[filter_Lakers_Players_From_USA, 'player_height'].mean()
americanLakersMean
```

# Shortest non american Lakers player

```python
shortestNonAmericanLaker = df.loc[filter_Lakers_Players_Not_From_USA, 'player_height'].min()
shortestNonAmericanLaker
```

### Tallest non american Lakers player

```python
tallestNonAmericanLaker = df.loc[filter_Lakers_Players_Not_From_USA, 'player_height'].max()
tallestNonAmericanLaker
```

### Tallest american Lakers player

```python
tallestAmericanLaker = (df.loc[filter_Lakers_Players_From_USA, 'player_height']).max()
tallestAmericanLaker
```

### Shortest american Lakers player

```python
shortestAmericanLaker = (df.loc[filter_Lakers_Players_From_USA, 'player_height']).min()
shortestAmericanLaker
```

### Diagram of LA Lakers players average points by age

```python
import matplotlib.pyplot as plt

x = df.loc[filter_Lakers_Players_Not_From_USA, 'age']
y = df.loc[filter_Lakers_Players_Not_From_USA, 'pts']
plt.xlabel('Player age')
plt.ylabel('Average points per game')
plt.bar(x,y)
plt.show()
```
